**Applying label encoder**

```python
labelled_data.select_dtypes(include='object')
```

```python
from sklearn import preprocessing
```

```python
for col in labelled_data.columns:
    if labelled_data.dtypes[col] == "object":
        le = preprocessing.LabelEncoder()
        le.fit(labelled_data[col])
        labelled_data[col] = le.transform(labelled_data[col])
```

```python
labelled_data[['X0','X1','X2','X3','X4','X5','X6','X8']]
```

```python
final_train_data=labelled_data
```

```python
for col in labelled_data1.columns:
    if labelled_data1.dtypes[col] == "object":
        le = preprocessing.LabelEncoder()
        le.fit(labelled_data1[col])
        labelled_data1[col] = le.transform(labelled_data1[col])
```

```python
final_test_data=labelled_data1
```

```python
from sklearn.preprocessing import StandardScaler  # to standardize the features
from sklearn.decomposition import PCA  # to apply PCA
import seaborn as sns  # to plot the heat maps
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(final_train_data,train_target,test_size=0.25, random_state=0)
```

```python
pca = PCA(n_components = 3)
x_train=pca.fit_transform(x_train)
x_test=pca.fit_transform(x_test)
final_test_data = pca.transform(final_test_data)
final_train_data = pca.transform(final_train_data)
```
